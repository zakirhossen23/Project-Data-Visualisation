#DataFrame 1 for daily visitor data

1: Reading csv

2: Converting into **DataFrame**


---



```python
import pandas as pd
from matplotlib.patches import Patch

visitordata=pd.read_csv("https://raw.githubusercontent.com/ChrisWalshaw/DataViz/master/Data/001145733/VenueDailyVisitors.csv")

df=pd.DataFrame(visitordata)

df
```

1: Compressing Date. (eg. 30 days to 1 month)

2: Resampling dataframe with Mean for all the Venue Ids.

3: Converting Date to index(one row for each date)


---



```python
df.index = pd.to_datetime(df['Date'])
df = df.resample('M').mean()
df = (df.set_index(df.index)
         .rename_axis("").T)
df
```

 **Visualizations** 

---



**Normal Seasonality visualisation of Mean by *Monthly* view**

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(df)  

plt.show()
```

# DataFrame 2 contain Summary Data

**Gathering all Datas into one DataFrame**

---



```python
VenueAgeData=pd.read_csv("https://raw.githubusercontent.com/ChrisWalshaw/DataViz/master/Data/001145733/VenueAge.csv")
VenueDistanceData=pd.read_csv("https://raw.githubusercontent.com/ChrisWalshaw/DataViz/master/Data/001145733/VenueDistance.csv")
VenueDurationData=pd.read_csv("https://raw.githubusercontent.com/ChrisWalshaw/DataViz/master/Data/001145733/VenueDuration.csv")
VenueGenderData=pd.read_csv("https://raw.githubusercontent.com/ChrisWalshaw/DataViz/master/Data/001145733/VenueGender.csv")
VenueSpendData=pd.read_csv("https://raw.githubusercontent.com/ChrisWalshaw/DataViz/master/Data/001145733/VenueSpend.csv")

df2 = pd.DataFrame(VenueAgeData.pivot_table('Avg age (yrs)',[],'Id')) #Age Data
visitordata2 = visitordata.iloc[0:, 1:]
# Converting the index as date
visitordata2.index = visitordata2.index
visitordata2=visitordata2.resample('A').mean()
visitordata3 = visitordata2.reset_index().iloc[0:, 1:]
visitordata5 = pd.pivot_table(visitordata3,columns='Avg visit (yrs)').reset_index()
visitordata5.columns=['Id',"Avg visit (yrs)"]
df2 = df2.append(visitordata5.pivot_table('Avg visit (yrs)',[],'Id')) #Visitor Data

df2 = df2.append(VenueDistanceData.pivot_table('Max travel distance (mls)',[],'Id'))#Distance Data
df2 = df2.append(VenueDurationData.pivot_table('Avg visit duration (mins)',[],'Id'))#Duration Data
df2 = df2.append(VenueGenderData.pivot_table('Proportion Female (%)',[],'Id'))#Gender Data
df2 = df2.append(VenueSpendData.pivot_table('Avg spend (£)',[],'Id'))#Spend Data

df2
```

**Maximum status for all the Venue ID**

---



```python
Maximum =df2.max(axis=1).to_frame()
Maximum = (Maximum.set_index(Maximum.index)
        .rename_axis(""))
MaximumVenue = df2.idxmax(axis=1).to_frame()
Maximum.head()
MaximumVenue.head()
```

 **Visualizations**

---



```python
ax = Maximum.plot(kind='bar',figsize=[8,8],legend=False)
plt.xticks(rotation=30, horizontalalignment="center")

x_offset = -0.28
y_offset = 1
for x in range(5):
    p =ax.patches[x]
    b = p.get_bbox()
    val = "{} - {}".format(MaximumVenue[0].values.tolist()[x], int(b.y1 ))
    ax.annotate(val, ((b.x0 + b.x1)/2 + x_offset, b.y1 + y_offset))
```
